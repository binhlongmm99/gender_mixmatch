```python
import  os

import numpy as np

import torch
import torch.utils.data as data
from torch.utils.data import Dataset, DataLoader

import torchvision
from torchvision import datasets, models, transforms
```

```python
INPUT_DIR_PATH = r"E:\icomm\data\domain_hawkice_damdev"
transform_train = transforms.Compose([
        transforms.RandomResizedCrop(32),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
    ])

train_labeled_path = os.path.join(INPUT_DIR_PATH, 'train')
dataset_labeled_train = datasets.ImageFolder(train_labeled_path, transform = transform_train)

```

```python
print(len(dataset_labeled_train))
print(dataset_labeled_train.imgs)
print(dataset_labeled_train.targets)
```

```python
print(dataset_labeled_train.classes)
```

```python
labeled_trainloader = data.DataLoader(dataset_labeled_train, batch_size=32, 
                                        #   shuffle=True, num_workers=0, drop_last=True
                                          )
```

```python
print(labeled_trainloader.__dict__)
print(len(labeled_trainloader))
```

```python
data          = next(iter(labeled_trainloader))
print(len(data))
actual_labels = data[1]
print(len(actual_labels))

print(actual_labels)
```
